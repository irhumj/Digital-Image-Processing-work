```python
import cv2
import numpy as np

# Load the image
img = cv2.imread(r'C:\Users\DELL\Pictures\download.jpg')

# Convert the image to a numpy array for easy manipulation
height, width, channels = img.shape

# Define the brightness adjustment factor (increase intensity by this value)
brightness_factor = 50  # you can adjust this value

# Loop over each pixel in the image
for i in range(height):
    for j in range(width):
        for c in range(channels):
            # Increase the intensity of each pixel by the brightness_factor
            new_value = img[i, j, c] + brightness_factor
            
            # Ensure the new value is within the valid range [0, 255]
            if new_value > 255:
                new_value = 255
            elif new_value < 0:
                new_value = 0

            # Assign the new value back to the image
            img[i, j, c] = new_value

# Display the brightened image
cv2.imshow('Brightened Image', img)
cv2.waitKey(0)

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread(r'C:\Users\DELL\Pictures\download.jpg')

# Convert to the range [0, 1] by dividing by 255
img = img / 255.0

# Define the gamma value
gamma = 2.2  # You can adjust this value to increase/decrease brightness

# Perform gamma correction using the formula output_pixel = input_pixel^(1/gamma)
gamma_corrected = np.power(img, 1 / gamma)

# Convert back to the range [0, 255] and to uint8
gamma_corrected = np.uint8(gamma_corrected * 255)
img_uint8 = np.uint8(img * 255)

# Convert BGR to RGB for displaying in Matplotlib
img_rgb = cv2.cvtColor(img_uint8, cv2.COLOR_BGR2RGB)
gamma_corrected_rgb = cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2RGB)

# Display the original and gamma corrected images side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected_rgb)
plt.title(f'Gamma Corrected (Gamma = {gamma})')
plt.axis('off')

plt.show()

```

```python
import cv2
import numpy as np

# 1. Load the foreground (green screen) image and background images
foreground = cv2.imread(r'foreground.jpg')  # Green screen image
background1 = cv2.imread(r'background1.jpg')
background2 = cv2.imread(r'background2.jpg')
background3 = cv2.imread(r'background3.jpg')

# Resize backgrounds to match the foreground image size
background1 = cv2.resize(background1, (foreground.shape[1], foreground.shape[0]))
background2 = cv2.resize(background2, (foreground.shape[1], foreground.shape[0]))
background3 = cv2.resize(background3, (foreground.shape[1], foreground.shape[0]))

# 2. Convert the foreground image from BGR to HSV
hsv_foreground = cv2.cvtColor(foreground, cv2.COLOR_BGR2HSV)

# 3. Define the HSV range for green
lower_green = np.array([35, 100, 100])
upper_green = np.array([85, 255, 255])

# 4. Create a mask to detect the green color
mask = cv2.inRange(hsv_foreground, lower_green, upper_green)

# 5. Invert the mask to get the object
mask_inv = cv2.bitwise_not(mask)

# 6. Extract the object (foreground) without the green background
fg_part = cv2.bitwise_and(foreground, foreground, mask=mask_inv)

# Function to combine foreground with any background
def combine_with_background(background):
    # 7. Extract the corresponding region from the background image
    bg_part = cv2.bitwise_and(background, background, mask=mask)
    
    # 8. Combine the foreground and background
    final_image = cv2.add(fg_part, bg_part)
    
    return final_image

# Apply the combination for each background
final_image1 = combine_with_background(background1)
final_image2 = combine_with_background(background2)
final_image3 = combine_with_background(background3)

# 9. Display the final results
cv2.imshow("Final Image with Background 1", final_image1)
cv2.imshow("Final Image with Background 2", final_image2)
cv2.imshow("Final Image with Background 3", final_image3)

# Press any key to close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```
