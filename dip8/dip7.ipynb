```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    if x <= 0.25:
        return 1 - x
    elif x <= 0.5: 
        return 1
    else:
        return x  
       
xx = np.linspace(0, 1, 500)  
y = np.array([f(x) for x in xx])

plt.plot(xx, y, label='f(x)')  # Corrected: using 'xx' instead of 'x'
plt.title("Graph of f(x) for Image Intensities")
plt.xlabel("x (Image Intensity)")
plt.ylabel("f(x)")
plt.legend()
plt.show()

```

```python
image=cv2.imread(r'C:\Users\DELL\Videos\images.jpg')
def transform_pixel_value(x):
    if x <= 0.25:
        return 1 - x
    elif x <= 0.5:
        return 1
    else:
        return x

def apply_transformation(image):
    image_normalized = image / 255.0
    transformed_image = np.vectorize(transform_pixel_value)(image_normalized)
    
    transformed_image = (transformed_image * 255).astype(np.uint8)
    return transformed_image

transformed_image = apply_transformation(image)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Transformed Image")
plt.imshow(transformed_image, cmap='gray')

plt.show()
```
