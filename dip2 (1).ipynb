```python
import cv2
import os


def resize_image(image, width=None, height=None):
    if width is None and height is None:
        return image
    return cv2.resize(image, (width, height))

#  Image in the Same Folder
same_folder_image = cv2.imread(r'C:\Users\DELL\Videos\irhum.jpg')
if same_folder_image is not None:
    print("Image loaded from the same folder successfully.")
    same_folder_image = resize_image(same_folder_image, width=400, height=300)
else:
    print("Failed to load image from the same folder.")

#  Image in a Different Folder (images folder)
different_folder_image = cv2.imread(r'C:\Users\DELL\Videos\irhum.jpg')
if different_folder_image is not None:
    print("Image loaded from the 'images' folder successfully.")
    
    different_folder_image = resize_image(different_folder_image, width=400, height=300)
else:
    print("Failed to load image from the 'images' folder.")

# 3. Image Using an Absolute Path (data folder)
absolute_path = os.path.abspath(r'C:\Users\DELL\Videos\irhum.jpg')
absolute_path_image = cv2.imread(absolute_path)
if absolute_path_image is not None:
    print("Image loaded using an absolute path successfully.")
    absolute_path_image = resize_image(absolute_path_image, width=400, height=300)
else:
    print("Failed to load image using an absolute path.")

# 4. Image Using a Relative Path (data folder)
relative_path_image = cv2.imread(r'C:\Users\DELL\Videos\irhum.jpg')
if relative_path_image is not None:
    print("Image loaded using a relative path successfully.")
    relative_path_image = resize_image(relative_path_image, width=400, height=300)
else:
    print("Failed to load image using a relative path.")

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import Image
import seaborn as sns
import matplotlib.pyplot as plt  

# File path of the image
image_path = r'C:\Users\DELL\Videos\irhum.jpg'

#  Load and Display Image using PIL 
image_pil = Image.open(image_path)
image_pil.show()  

#  Load and Display Image using OpenCV
image_cv = cv2.imread(image_path)
image_cv_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB) 
cv2.imshow('Image using OpenCV', image_cv)  
cv2.waitKey(0)
cv2.destroyAllWindows()

#  Display Image using Seaborn (indirectly through Matplotlib)
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

# Create a figure for each style
for style in styles:
    sns.set_style(style)  
    plt.figure(figsize=(8, 6))   
sns.heatmap(image_cv_rgb[:, :, 0], cmap='viridis', cbar=False)  
plt.axis('off')  # Turn off axis
plt.title("Image displayed with Seaborn")
plt.show()
```
